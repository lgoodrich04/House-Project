```python
# Converting Yes to 1 and No to 0 using map function

# Create a mapping dictionary for converting 'yes' to 1 and 'no' to 0
yes_no_map = {'yes': 1, 'no': 0}

# Use the map function to convert specific columns
# For each column, the map() function replaces values based on the mapping dictionary

# Convert 'mainroad' column
housing['mainroad'] = housing['mainroad'].map(yes_no_map)

# Convert 'guestroom' column
housing['guestroom'] = housing['guestroom'].map(yes_no_map)

# Convert 'basement' column
housing['basement'] = housing['basement'].map(yes_no_map)

# Convert 'hotwaterheating' column
housing['hotwaterheating'] = housing['hotwaterheating'].map(yes_no_map)

# Convert 'airconditioning' column
housing['airconditioning'] = housing['airconditioning'].map(yes_no_map)

# Convert 'prefarea' column
housing['prefarea'] = housing['prefarea'].map(yes_no_map)

# Display the DataFrame after conversion
print("DataFrame after converting 'yes' to 1 and 'no' to 0:")
housing.head()
```

```python
housing["furnishingstatus"].unique()
```

```python
status=pd.get_dummies(housing["furnishingstatus"])
status.head()
```

```python
housing=pd.concat([housing,status],axis=1)
housing.head()
```

```python
housing.drop(["furnishingstatus"],axis=1,inplace=True)
housing.head()
```

```python
for i in range(len(housing['furnished'])):
    housing['furnished'][i] = np.int64(housing['furnished'][i])
```

```python
for i in range(len(housing['semi-furnished'])):
    housing['semi-furnished'][i] = np.int64(housing['semi-furnished'][i])
```

```python
for i in range(len(housing['unfurnished'])):
    housing['unfurnished'][i] = np.int64(housing['unfurnished'][i])
```

```python
# Defining a normalization function using the formula: (x - min(x)) / (max(x) - min(x))

def normalize(x):
    """
    Normalizes the input array using the formula: (x - min(x)) / (max(x) - min(x))

    Parameters:
    x (array-like): Input array to be normalized

    Returns:
    array-like: Normalized array
    """
    return (x - np.min(x)) / (np.max(x) -  np.min(x))

# Applying the normalize() function to all columns using the 'apply()' function

# Note: The 'apply()' function applies the 'normalize()' function to each column independently
housing = housing.apply(normalize)

# Display the normalized 'housing' DataFrame
print("Normalized 'housing' DataFrame:")
housing.head()
```

```python
from sklearn.model_selection import train_test_split

X = housing[['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad',
             'guestroom', 'basement', 'hotwaterheating', 'airconditioning',
             'parking', 'prefarea', 'semi-furnished', 'unfurnished']]

y = housing['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
from sklearn.linear_model import LinearRegression

# Defining and fitting the Linear Regression model

# Creating an instance of the LinearRegression class
lm_2 = LinearRegression()

# Fitting the model to the training data
lm_2.fit(X_train, y_train)

# Displaying the trained model
print("Trained Linear Regression model:", lm_2)
```

```python
y_pred = lm_2.predict(X_test)
print(y_pred)
```
